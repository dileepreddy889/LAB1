```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC, LinearSVC
```

```
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/lab1/glass.csv') #Reading the data from glass.csv
```

```
X = data.drop(["Type"], axis=1)
Y = data["Type"]
```

```
Label_Encoder = LabelEncoder()
data["Type"] = Label_Encoder.fit_transform(data["Type"])  #label Encoderr to convert non numeric to numeric feature
```

```
data=data.apply(lambda x: x.fillna(x.mean()),axis=0)   #Filling null values with mean using lamda function
print(data.isnull().sum())
```

```
X_train, X_test, Y_train, y_test= train_test_split(X, Y, test_size=0.4, random_state=0) #splitting train and test data
```

```
gnb = GaussianNB()

k = gnb.fit(X_train, Y_train).predict(X_test)
gnb.fit(X_train, Y_train)
k = gnb.predict(X_test)
Accuracy = gnb.score(X_test, y_test)
print("Gaussian accuracy is:",Accuracy)
```

```
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)
k = knn.predict(X_test)
Accuracy = knn.score(X_test, y_test)
print("Knn accuracy is:",Accuracy)
```

```
svc = SVC()
svc.fit(X_train, Y_train)
k = svc.predict(X_test)
Accuracy = svc.score(X_test, y_test)
print("svm accuracy is:", Accuracy)
```
